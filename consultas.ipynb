```python
import duckdb

# Conectamos a nuestro motor local
con = duckdb.connect()

# Escribimos nuestra consulta SQL (igual que en BigQuery)
# Excluimos PitInTime y PitOutTime que tienen tipo TIME WITH TIME ZONE no soportado
query = """
SELECT * EXCLUDE (PitInTime, PitOutTime) FROM 'data/silver/silver_laps.parquet'
LIMIT 100
"""

# Ejecutamos y mostramos la tabla visual
con.execute(query).df()
```

```python
# Primero, veamos el esquema del archivo para identificar las columnas
schema_query = """
DESCRIBE 'data/silver/silver_laps.parquet'
"""
schema = con.execute(schema_query).df()
print("Esquema del archivo:")
print(schema)
```
